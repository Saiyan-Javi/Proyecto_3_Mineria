```python
import pandas as pd

# Leer el archivo como texto
with open("Merged file.csv", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Mostrar las primeras líneas para inspeccionar el contenido
for line in lines[:20]:
    print(line)

import pandas as pd

# Leer el archivo como texto
with open("Merged file.csv", "r", encoding="utf-8") as file:
    lines = file.readlines()

# Encontrar el inicio y fin de la tabla
start_index = lines.index('" Divorcios por grupos de edad del esposo, según grupos de edad de la esposa, al momento de la sentencia, año 2015 "\n')
end_index = lines.index('" Divorcios por pueblo de pertenencia del hombre, según pueblo de pertenencia de la mujer, año 2015 "\n')

# Extraer las líneas de la tabla
table_lines = lines[start_index:end_index]

# Limpiar las líneas
cleaned_lines = [line.strip() for line in table_lines if line.strip()]

# Convertir a DataFrame
df = pd.DataFrame([line.split(",") for line in cleaned_lines])


# Especificar la ruta del archivo CSV
csv_file_path = "Merged file.csv"

# Cargar el archivo CSV
try:
    df = pd.read_csv(csv_file_path)
    print("Archivo CSV cargado correctamente.")
except Exception as e:
    print(f"Error al cargar el archivo CSV: {e}")
    exit()  # Detiene el programa si hay un error

# Descripción general del conjunto de datos
def describe_dataset(df):
    print(f"\nDescripción del DataFrame:")
    print(f"Número de filas: {df.shape[0]}")
    print(f"Número de columnas: {df.shape[1]}")
    print("\nPrimeras filas del DataFrame:")
    print(df.head())
    print("\nInformación del DataFrame:")
    print(df.info())
    print("\nResumen estadístico de las variables numéricas:")
    print(df.describe())
    print("\nResumen de las variables categóricas:")
    print(df.describe(include=['object']))

# Llamar a la función describe_dataset solo si df está definido
if 'df' in globals():  # Verifica si df está definido
    describe_dataset(df)
else:
    print("Error: No se pudo cargar el DataFrame.")

# Significado y tipo de cada variable
def variable_types(df):
    print(f"\nTipos de variables en el DataFrame:")
    print(df.dtypes)

if 'df' in globals():
    variable_types(df)

# Cantidad de variables y observaciones
def count_variables_observations(df):
    print(f"\nCantidad de variables: {df.shape[1]}")
    print(f"Cantidad de observaciones: {df.shape[0]}")

if 'df' in globals():
    count_variables_observations(df)

# Exploración de variables numéricas
def explore_numeric_variables(df):
    print(f"\nExploración de variables numéricas:")
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numeric_columns:
        print(f"\nVariable: {col}")
        print(f"Media: {df[col].mean()}")
        print(f"Mediana: {df[col].median()}")
        print(f"Desviación estándar: {df[col].std()}")
        print(f"Mínimo: {df[col].min()}")
        print(f"Máximo: {df[col].max()}")
        print(f"Distribución:")
        print(df[col].value_counts())

if 'df' in globals():
    explore_numeric_variables(df)

# Exploración de variables categóricas
def explore_categorical_variables(df):
    print(f"\nExploración de variables categóricas:")
    categorical_columns = df.select_dtypes(include=['object']).columns
    for col in categorical_columns:
        print(f"\nVariable: {col}")
        print(f"Frecuencia de categorías:")
        print(df[col].value_counts())

if 'df' in globals():
    explore_categorical_variables(df)

# Relaciones entre las variables
def explore_relationships(df):
    print(f"\nRelaciones entre las variables:")
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    if len(numeric_columns) > 1:
        print("Matriz de correlación:")
        print(df[numeric_columns].corr())
    else:
        print("No hay suficientes variables numéricas para calcular la correlación.")

if 'df' in globals():
    explore_relationships(df)

```
